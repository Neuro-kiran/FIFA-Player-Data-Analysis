```python
f.columns = f.columns.str.replace(' ', '_')
print(f.head(3))
```

```python
f = f.drop('ID', axis=1)
print(f.head(3))
```

```python
barcelona_data = f[f['Club'] == 'FC Barcelona']
print(barcelona_data)
```

```python
# Filter the players with age > 30
players_over_30 = f[f['Age'] > 30]

# Count the number of players over 30
count_players_over_30 = len(players_over_30)

# Display the players over 30 and the count
print("Players over 30 years old:")
print(players_over_30)
print("\nNumber of players over 30:", count_players_over_30)
```

```python
# Count the number of players for each preferred foot
foot_counts = f['Preferred_Foot'].value_counts()

# Create a pie plot
plt.figure(figsize=(8, 8))
plt.pie(foot_counts, labels=foot_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Preferred Foot of Players')

# Show the pie plot
plt.show()
```

```python
# Filter the players from Germany with age < 20
german_players_under_20 = f[(f['Nationality'] == 'Germany') & (f['Age'] < 20)]

# Display the selected players
print(german_players_under_20)
```

```python
# Filter the players with Value between 50,000 and 110,000
selected_players = f[(f['Value'] > 50000) & (f['Value'] < 110000)]

# Display the selected players
print(selected_players)
```

```python
# Filter the players with Skill Moves equal to 5.0
selected_players = f[f['Skill_Moves'] == 5.0]

# Display the selected players
print(selected_players)
```

```python
# Filter the record for Neymar Jr
neymar_record = f[f['Name'] == 'Neymar Jr']

# Display the record for Neymar Jr
print(neymar_record)
```

```python
# Group the data by 'Nationality' (country) and calculate the mean height and weight
country_avg_height_weight = f.groupby('Nationality')[['Height', 'Weight']].mean()

# Display the country-wise average height and weight
print(country_avg_height_weight)
```
